```python
sns.pairplot(df, vars=['latitude','Speed_Mph', 'Longitude', 'Gforce_clean'],hue='Type_Main')
plt.show() #HERE DISTRIBUTION OF EACH FEATURE IS SHOWN
```

```python
#ALSO WE CAN CORRELATE WITH THE SEABORN'S HEATMAPS
sns.heatmap(df_coor,annot=True)
```
